### DFS/BFS

- n개의 음이 아닌 정수들이 있음
- 이 정수들을 **순서를 바꾸지 않고 적절히 더하거나 빼서 타겟 넘버**를 만들려고 함
- numbers: 사용할 수 있는 숫자가 담긴 배열
- target: 타겟 넘버
- 숫자를 **적절히 더하고 빼서 타겟 넘버를 만드는 방법의 수**를 return

**제한사항**

- 주어지는 숫자의 개수는 2개 이상 20개 이하
- 각 숫자는 1 이상 50 이하인 자연수
- 타겟 넘버는 1 이상 1000 이하인 자연수

```python
"""
입출력 예시)

numbers = [1, 1, 1, 1, 1], target = 3 -> 5
numbers = [4, 1, 2, 1], target = 4 -> 2
"""
```

```python
# numbers = [1, 1, 1, 1, 1], target = 3 -> 5

# -1+1+1+1+1 = 3
# +1-1+1+1+1 = 3
# +1+1-1+1+1 = 3
# +1+1+1-1+1 = 3
# +1+1+1+1-1 = 3

# => 5가지 
```

```python
# numbers = [4, 1, 2, 1], target = 4 -> 2

# +4+1-2+1 = 4
# +4-1+2-1 = 4

# => 2
```

### 통과)

```python
def solution(numbers, target):
    answer = 0
    q = [[numbers[0], 0], [-1*numbers[0], 0]] # [+정수, 인덱스], [-정수, 인덱스]
    while q: 
        temp, idx = q.pop()
        # 다음 인덱스에 위치한 값으로 탐색
        idx += 1
        if idx < len(numbers):
            q.append([temp+numbers[idx], idx])
            q.append([temp-numbers[idx], idx])
        else:
            if temp == target:
                answer += 1
    return answer
```

```python
numbers = [1, 1, 1, 1, 1]
target = 3 
solution(numbers, target)
```

```python
numbers = [4, 1, 2, 1]
target = 4
solution(numbers, target)
```

```python
def solution(numbers, target):
    answer = 0 # 방법의 수
    q = [[numbers[0], 0], [-1*numbers[0], 0]]
    print('시작 q:', q, '\n')
    while q:
        temp, idx = q.pop()
        print('temp:', temp)
        print('idx:', idx)
        # 다음 인덱스에 위치한 값으로 탐색 
        idx += 1
        if idx < len(numbers):
            print('다음 인덱스에 위치한 numbers 원소의 + 값:', numbers[idx])
            print('다음 인덱스에 위치한 numbers 원소의 - 값:', -1* numbers[idx])
            q.append([temp+numbers[idx], idx])
            q.append([temp-numbers[idx], idx])
            print('갱신된 q:', q, '\n')
        else:
            if temp == target:
                answer += 1
    return answer
```

```python
numbers = [1, 1, 1, 1, 1]
target = 3 
solution(numbers, target)
```
